# **Modelo Dengue Grave em Crianças e Adultos** <h1>
Baseado no modelo de **[Adimy 2020](https://link.springer.com/article/10.1007/s11538-020-00699-x)** e no conhecimento da resposta imune contra o vírus da dengue sob a hipótese de aprimoramento dependente de anticorpos (ADE), um modelo de equações diferenciais ordinárias não-lineares é proposto para investigar a interação entre o vírus da dengue, as células alvo (não infectadas e infectadas) e os anticorpos (neutralizantes e não neutralizantes) durante a segunda infecção (heteróloga) por DENV.

No modelo, os anticorpos de reação cruzada (não neutralizantes) $A_1$ são produzidos pelas células plasmáticas a uma taxa constante $\Lambda_1$, e morrem a uma taxa de morte natural  $\gamma_A$. Essa população também aumenta proporcionalmente à presença de células infectadas $Y$ a uma taxa $ \alpha_1$ sendo consumida quando ligada ao vírus $V$ a uma taxa $\eta_1$ (complexo vírus-anticorpo). Os anticorpos neutralizantes $A_2$ são gerados devido à presença das células infectadas a uma taxa $\alpha_2$, morrem a uma taxa de mortalidade natural $\gamma_A$, e diminuem quando ligados ao vírus (complexo vírus-anticorpo) a uma taxa $\eta_2$.

A população do vírus livre $V$ é produzida a uma taxa $\Phi$ devido à presença de células infectadas $Y$. Quando os anticorpos $A_{1}$ e $A_{2}$, se ligam ao receptor do vírus $V$, o vírus não é mais livre a uma taxa $\xi_1$ e $\xi_2$, respectivamente (complexo vírus-anticorpo). Além disso, a população do vírus $V$ tem uma taxa de morte natural $\gamma_V$.

As células-alvo suscetíveis $X$ são produzidas a uma taxa $\Omega$, têm taxa de morte natural $\gamma_X$ e se tornam infectadas pelo vírus $V$ a taxas $\beta_1$ e $\beta_2$.  As células infectadas $Y$ são geradas a partir da infecção de células suscetíveis $X$ e morrem a uma taxa $\gamma_Y \ge \gamma_X$. A interação entre vírus, célula saudável e anticorpo é modelada por funções que serão descritas a seguir.


A função de neutralização é definida a partir da reação química **[Santillan (2008)](https://www.cambridge.org/core/journals/mathematical-modelling-of-natural-phenomena/article/abs/on-the-use-of-the-hill-functions-in-mathematical-modelsof-gene-regulatory-networks/985926A8D13CB5E075C496F29642094F)**:
	     $$ R + nL \rightleftharpoons R_{nL}, $$
       em que $n$ moléculas do ligante $L$ se ligam simultaneamente ao receptor $R$.

Quando a reação química entra em equilíbrio:
$$ k^{-}[R][L]^n = k^{+}[R_{nL}] \quad \Rightarrow \quad K  = \dfrac{[reagentes]}{[produto]}= \dfrac{[R][L]^n}{[R_{nL}]}, $$ sendo $K =  \dfrac{k^{+}}{k^{-}},$ a constante de dissociação da reação, e $[\cdot]$ a concentração de espécie química.

Dado que o número de receptores são constantes (e diferente para cada vírus) temos que:
  $$[R] + \left[R_{nL}\right]= \left[R_{\text{total}}\right],$$
e podemos escrever as frações das moléculas ocupadas como sendo
  
  $$N = \dfrac{[R_{nL}]}{[R] + [R_{nL}]} = \dfrac{\dfrac{[R][L]^{n}}{K}}{[R] + \dfrac{[R][L]^{n}}{K}} = \dfrac{[L]^{n}}{K + [L]^{n}}.$$

Supondo que a concentração de anticorpos neutralizantes $A_2$ é diretamente proporcional a concentração de moléculas $L$, ou seja,
$$ A_2 = k[L]$$ e substituindo a expressão de $A_{2}$ na equação \eqref{neutraN}, obtemos:
$$ N = \dfrac{\left(\dfrac{A_2}{k}\right)^{n}}{K + \left(\dfrac{A_2}{k}\right)^{n}} = \dfrac{A_2^{n}}{k^{n}K + A_2^{n}}.$$

  Definindo
	 	$$\theta_{2} = k\sqrt[n]{K},$$ obtemos:
	$$ N(A_{2}) = \dfrac{A_{2}^{n}}{\theta_{2}^{n} + A_{2}^{n}},$$
      e podemos interpretar $N$ como sendo a probabilidade de um vírus ser neutralizado pelo anticorpo $A_2$. Por outro lado, a probabilidade de um vírus não ser neutralizado pelo anticorpo $A_2$ é dada por:
$$ N_{0} = 1 - N(A_{2}) = \dfrac{\theta_{2}^{n}}{\theta_{2}^{n} + A_2^{n}}.$$

Entretanto, nem todos os anticorpos que se ligam ao vírus conseguem neutralizá-lo, devido à baixa afinidade. Para lidar com este fenômeno, consideramos um estado intermediário de reação química no qual as frações das moléculas ocupadas pelos anticorpos pré-existentes $A_1$ são modeladas pela função de Hill, portanto:
	 	$$ E= \dfrac{[R_{pL}]}{[R] + [R_{nL}]}  = \dfrac{[L]^{p}}{K + [L]^{n}} = \dfrac{k^{n-p}A_{1}^{p}}{k^{n}K + A_{1}^{n}},$$  com $A_1 = k[L]$.

Definindo arbitrariamente $\theta_{1} = k\sqrt[n]{K}$ e $C = K^{\frac{p-n}{n}}=1$ (esses valores serão multiplicados depois pelas taxas de infecção), $E$ é dado por:
  $$ E =  \dfrac{\theta_1^{n-p}A_{1}^{p}}{\theta_{1}^{n} + A_{1}^{n}} \quad  \text{com} \quad 0 \leq p \leq n. $$


Portanto, a interação entre vírus $V$, célula saudável $X$ e anticorpo é modelada por é $E(A_1)V$ (aprimoramento da infecção) ou $N(A_2)$ (neutralização da infecção).


---


 O modelo é descrito por:
$$
		\begin{cases}
			\dfrac{dA_{1}}{d t} & =\Lambda + \alpha_{1}Y - \gamma_{A}A_{1} - \eta_{1}A_{1} V \\
			\dfrac{dA_{2}}{d t} & = \alpha_{2}Y - \gamma_{A}A_{2} - \eta_{2}A_{2} V \\
			\dfrac{d X}{d t} & = \Omega - \gamma_{X}X -\beta_{1}E(A_{1})VX -\beta_{2}N_{0}(A_{2})VX\\
			\dfrac{dY}{d t} & = \beta_{1}E(A_{1})VX + \beta_{2}N_{0}(A_{2})VX- \gamma_{Y}Y\\
			\dfrac{dV}{d t} & = \Phi Y - \xi_{1} A_{1}V -\xi_{2} A_{2}V - \gamma_{V} V
		\end{cases}
$$
  
  e podemos identificar as funções de aprimoramento $E$ e neutralização de vírus $N$  
  
$$
		E(A_{1}) = \dfrac{\theta_{1}^{n-p}A_{1}^{p}}{\theta_{1}^{n} + A_{1}^{n}} \quad \text{e} \quad N(A_{2}) =  \dfrac{A_2^{n}}{\theta_{2}^{n} + A_2^{n}}\quad \mbox{com}\quad   N_0(A_2) = 1 - N (A_2).
$$

# 2. Método: Runge-Kutta de 4ª ordem - Clássico

Para entender como se aplica o método, consideremos o problema de valor inicial (PVI) genérico:
$$
\begin{cases}
y_{0} = f(x, y)\\
y(x_{0})= y_{0}
\end{cases}
$$

Tomamos combinações de $f(x, y)$ em vários pontos do intervalo $x_{k} \leq x \leq x_{k+1}$ e
ajustamos os parâmetros para que nossa aproximação de $y_{k+1}$ coincida com a expansão de Taylor de ordem 4. Dessa forma,

$$y_{k+1} = y_{k} + \dfrac{h}{6}(K_{1} + 2(K_{2} + K_{3}) + K_{4}),$$
em que

$$  
\begin{cases}
K_{1} = f(x_{k}, y_{k}), \\
    K_{2} = f\left(x_{k} + \dfrac{h}{2}, y_{k} + \dfrac{hK_{1}}{2}\right),\\
    K_{3} = f\left(x_{k} + \dfrac{h}{2}, y_{k} + \dfrac{hK_{2}}{2}\right),\\   
    K_{4} = f\left(x_{k} + h, y_{k} + hK_{3}\right).
    \end{cases}
$$


```python
# Runge Kutta de 4ª ordem - Clássico
def RungeKutta4(f, s0, h, tf):
    n = int (tf/h)
    t = np.linspace(0, tf, n + 1) #tempo => 0 até tf
    s = np.array((n + 1)*[s0])      #matriz com as variáveis de estados Ex.: A, X, Y, V => Matriz (4, tf)

    for i in range(n):
        # f now expects both t and u. We pass t[i] as the current time.
        k1 = f(t[i], s[i])
        k2 = f(t[i] + 0.5 * h, s[i] + 0.5 * h*k1)
        k3 = f(t[i] + 0.5 * h, s[i] + 0.5 * h*k2)
        k4 = f(t[i] + h, s[i] + h*k3)

        s[i + 1] = s[i] + (h/6.0)*(k1 + 2*(k2 + k3) + k4)

    return t, s
```

# 3. Problema de Stiff  - Equações Diferenciais Rígidas

- O modelo de fato apresenta problemas de 'stiffness', logo, o Runge-Kutta de 4ª ordem pode não é adequado, pois exige passos de tempo muito pequenos para manter a estabilidade, tornando a simulação ineficiente.

- Para problemas de EDOs (Equações Diferenciais Ordinárias) **stiff**, o pacote mais recomendado em Python é o *scipy*.*integrate.solve_ivp*. Esta função é uma interface unificada para diferentes resolvedores de PVI (Problemas de Valor Inicial) e inclui métodos especializados para sistemas stiff.

Os métodos que você pode considerar dentro de solve_ivp são:

  - method='Radau': Este é um método implícito para sistemas stiff, conhecido por sua alta ordem de precisão e estabilidade.

  - method='BDF' (Backward Differentiation Formula): Outro método muito eficaz para sistemas stiff, especialmente quando as equações são muito stiff. Ele pertence à classe de métodos multipasso.

Usaremos o solve_ivp de forma semelhante ao RungeKutta4, mas com a vantagem de ter algoritmos que adaptam o tamanho do passo de forma mais inteligente para lidar com a rigidez do sistema.

## 3.1 Método RADAU

```python
from scipy.integrate import solve_ivp

def solve_ode_radau(f, s0, h, tf):
    # f: função que define as EDOs, f(t, u)
    # s0: vetor de estado inicial
    # h: tamanho do passo (usado para definir os pontos de t_eval)
    # tf: tempo final

    n_steps = int(tf / h)
    t_eval = np.linspace(0, tf, n_steps + 1)

    # solve_ivp espera f(t, y) onde t é o tempo e y é o vetor de estado.
    # Usando o método 'Radau' para problemas stiff.
    # rtol e atol são tolerâncias relativa e absoluta, boas para sistemas stiff.
    sol = solve_ivp(f, [0, tf], s0, method='Radau', t_eval=t_eval, rtol=1e-6, atol=1e-9)

    # solve_ivp retorna um objeto. sol.t são os pontos no tempo, sol.y é a matriz da solução.
    # sol.y possui as variáveis de estado em linhas e os pontos de tempo em colunas. Transponha.
    return sol.t, sol.y.T
```

## 3.2 Método LSODA

```python
import numpy as np
from scipy.integrate import solve_ivp

def solve_ode_lsoda(f, s0, h, tf):
    # f: função que define as EDOs, f(t, u)
    # s0: vetor de estado inicial
    # h: tamanho do passo (usado para definir os pontos de t_eval)
    # tf: tempo final

    n_steps = int(tf / h)
    t_eval = np.linspace(0, tf, n_steps + 1)

    # solve_ivp espera f(t, y) onde t é o tempo e y é o vetor de estado.
    # Usando o método 'LSODA' para problemas que podem alternar entre stiff e não-stiff.
    # rtol e atol são tolerâncias relativa e absoluta, boas para sistemas stiff.
    sol = solve_ivp(f, [0, tf], s0, method='LSODA', t_eval=t_eval, rtol=1e-6, atol=1e-9)

    # solve_ivp retorna um objeto. sol.t são os pontos no tempo, sol.y é a matriz da solução.
    # sol.y possui as variáveis de estado em linhas e os pontos de tempo em colunas. Transponha.
    return sol.t, sol.y.T
```

# 4. Definindo os Parâmetros do Modelo Geral

```python

gamma_x = np.log(2)/(25)  # taxa de mortalidade natural de  X, 30-60 [days] -- Mortality of susceptible cells
gamma_y = np.log(2)/(0.5) # taxa de mortalidade natural de Y, Mortality of infected cells
gamma_A = np.log(2)/(195) # taxa de mortalidade natural de A
gamma_v = np.log(2)*(24./2.7) # taxa de mortalidade natural de V, 2.7-17.2 hours -- Virus mortality

eta_1 = 6e-9   # taxa de consumo dos anticorpos A1 na presença do DENV, Consumption of A1 by V
eta_2 = 2e-9   # taxa de consumo dos anticorpos A2 na presença do DENV, Consumption of A2 by V
alpha_1 = 1.7e-4 # Production of A1 by memory cells
alpha_2 = 2.3e-4 # Production of A2 by memory cells

Omega = 2e5      # taxa de pordução constante de células
Phi = 5.3e4   # taxa de produção de vírus livres, produzidos por células infectadas
Lambda = 1 # Production of A1 may plasma B cells

xi_1 = 0.9 # taxa de consumo do vírus pelos anticorpos
xi_2 = 1 # taxa de consumo do vírus pelos anticorpos

theta_1 = 7e1
theta_2 = 2.5e2

beta_1 = 9e-9  # taxa de infecção
beta_2 = 1e-10  # taxa de infecção

n = 3
p = 1.2
```

# 5. Modelo Geral Implementado

```python
def ModeloGeral(t, u):  # Adicionado argumento 't' conforme exigido pelo solve_ivp
    A1, A2, X, Y, V = u

    # Garantir que A1 e A2 não sejam negativos para evitar erros com potências fracionárias - A1_safe e A2_safe
    A1_safe = np.maximum(1e-10, A1)  # Limitando A1 para um pequeno valor positivo
    A2_safe = np.maximum(1e-10, A2)  # Limitando A2 para um pequeno valor positivo

    # Definir E(A1) e N0(A2) com base nas equações do modelo, usando valores seguros
    E_A1 = ((theta_1**(n-p)) * (A1_safe**p)) / ((theta_1**n) + (A1_safe**n))

    # N0(A2) = 1 - N(A2) = 1 - (A2**n / (theta_2**n + A2**n)) = (theta_2**n) / (theta_2**n + A2**n)
    N0_A2 = (theta_2**n) / (theta_2**n + A2_safe**n)

    dA1 = Lambda + alpha_1*Y - gamma_A*A1 - eta_1*A1*V
    dA2 = alpha_2*Y - gamma_A*A2 - eta_2*A2*V
    dX  = Omega - gamma_x*X - beta_1*E_A1*X*V - beta_2*N0_A2*X*V
    dY  = beta_1*E_A1*X*V + beta_2*N0_A2*X*V - gamma_y*Y
    dV  = Phi*Y  - xi_1*A1*V  - xi_2*A2*V  - gamma_v*V

    return np.array([dA1, dA2, dX, dY, dV])
```

# Discussões:

1. **A influência de $E(A_1)$ e $N_0(A_2)$ na dinâmica viral.**

No modelo, tanto $E(A_1)$ quanto $N_0(A_2)$ desempenham papéis críticos na modulação do processo de infecção e, consequentemente, na dinâmica do vírus:

 - $E(A_1)$ (Função de Aprimoramento): Esta função modela o fenômeno de aprimoramento dependente de anticorpos (ADE). Ela está associada a anticorpos não neutralizantes $(A_1)$. Um valor mais alto de $E(A_1)$ significa que a presença desses anticorpos aprimora a taxa de infecção. Observando as equações das células suscetíveis $(X)$ e infectadas $(Y)$, o termo $E(A_1)$ multiplica diretamente o termo $V\cdot X$, o que significa que ele aumenta a taxa na qual as células suscetíveis $(X)$ se tornam infectadas pelo vírus $(V)$, impulsionando assim a produção de células infectadas $(Y)$. Um aumento nas células infectadas $(Y)$ leva diretamente a uma maior produção de vírus livre $(V)$, como visto na equação $\dfrac{dV}{dt}$ pelo termo ($\Phi\cdot Y$). Portanto, $E(A_1)$ promove a replicação viral e a progressão da doença.

  - $N_0(A_2)$ (Função de Não Neutralização): Esta função representa a probabilidade de que uma partícula viral não seja neutralizada por anticorpos neutralizantes $(A_2)$. Ela é definida como $1 - N(A_2)$, onde $N(A_2)$ é a probabilidade de neutralização. Um valor mais alto de $N_0(A_2)$ significa que menos partículas virais são neutralizadas, permitindo que mais delas infectem células suscetíveis. Semelhante a $E(A_1)$, $N_0(A_2)$ também multiplica o termo $V\cdot X$ nas equações $\dfrac{dX}{dt}$ e $\dfrac{dY}{dt}$, indicando que ele aumenta a taxa de infecção de células suscetíveis e a geração de células infectadas. Isso, por sua vez, contribui para uma maior carga viral. Portanto, $N_0(A_2)$ reflete inversamente a eficácia dos anticorpos neutralizantes; um $N_0(A_2)$ mais alto sugere um controle mais deficiente do vírus pelos anticorpos $A_2$.

Concluímos que, ambas as funções contribuem para a taxa de infecção e o subsequente aumento da população viral, mas através de mecanismos diferentes relacionados à interação entre o vírus e os diferentes tipos de anticorpos.


## Análise e explicação dos Efeitos de $\theta_1$ e $\theta_2$

**Análise dos Gráficos:**

1.  **Dinâmica Viral para Diferentes Valores de $\theta_1$ ($\theta_2$ Constante):**
    *   Observa-se que, à medida que $\theta_1$ aumenta, a população viral máxima (pico de viremia) tende a diminuir. Além disso, o tempo para atingir o pico pode ser ligeiramente atrasado ou a duração da viremia pode ser alterada.
    *   Valores mais baixos de $\theta_1$ resultam em picos virais mais altos e, potencialmente, em uma infecção mais severa.

2.  **Dinâmica Viral para Diferentes Valores de $\theta_{2}$ ($\theta_1$ Constante):**
    *   À medida que $\theta_{2}$ aumenta, a população viral máxima tende a aumentar. Isso significa que valores mais altos de $\theta_{2}$ estão associados a uma maior carga viral.
    *   Valores mais baixos de $\theta_{2}$ resultam em picos virais mais baixos, indicando um controle mais eficaz do vírus.

**Explicação dos Efeitos:**

Os parâmetros $\theta_1$ e $\theta_{2}$ desempenham papéis cruciais na modulação das funções $E(A_1)$ (aprimoramento) e $N0(A_2)$ (não neutralização), que, por sua vez, impactam diretamente a taxa de infecção e a replicação viral.

*   **Impacto de $\theta_1$ na $E(A_1)$:**
    *   A função de aprimoramento é dada por $E(A1) = (\theta_1^{n-p}\cdot A_1^p) / (\theta_1^n + A_1^n)$. No nosso caso, $n > p$ ($3 > 1.2$), então $n-p$ é positivo. Quando $A_1$ é muito menor que $\theta_1$, $E(A_1)$ se aproxima de $(A_1^p / \theta_1^p)$. Quando $A_1$ é muito maior que $\theta_1$, $E(A_1)$ se aproxima de $\theta_1^{n-p} / A_1^{n-p}$. O $\theta_1$ atua como uma constante de semi-saturação para a função $E(A_1)$. Um valor **menor** de $\theta_1$ significa que a função de aprimoramento $E(A_1)$ atinge seu máximo (ou se torna mais ativa) para concentrações mais baixas de anticorpos $A_1$. Isso implica que o aprimoramento da infecção ocorre mais facilmente e em níveis mais baixos de $A_1$, levando a uma taxa de infecção maior e, consequentemente, a uma maior carga viral. Por outro lado, um **maior** $\theta_1$ significa que $E(A_1)$ é menos sensível a $A_1$ ou que uma concentração muito maior de $A_1$ é necessária para atingir o aprimoramento máximo, o que pode mitigar o efeito de aprimoramento e reduzir a carga viral.

*   **Impacto de $\theta_{2}$ na $N0(A_2)$:**
    *   A função de não neutralização é $N_0(A_2) = \theta_2^n / (\theta_2^n + A_2^n)$. $\theta_{2}$ também atua como uma constante de semi-saturação para a função $N(A2)$ (neutralização).
    
    * Um valor **maior** de $\theta_{2}$ implica que a função $N_0(A_2)$ permanece alta (ou seja, a probabilidade de não neutralização é alta) para uma ampla gama de concentrações de $A_2$.
    
    * Isso significa que é mais difícil para os anticorpos neutralizantes $A_2$ neutralizarem o vírus, permitindo que mais vírus infectem as células. Consequentemente, uma maior taxa de infecção leva a uma maior replicação viral e a uma carga viral mais alta.
    
    * Inversamente, um **menor** $\theta_{2}$ faz com que $N_0(A_2)$ caia mais rapidamente com o aumento de $A_2$, significando que a neutralização é mais eficaz, o que resulta em uma menor carga viral.

**Conclusão:**

*   **$\theta_1$ (relacionado a $E(A_1)$ - aprimoramento):** Um $\theta_1$ menor facilita o aprimoramento da infecção, levando a uma maior carga viral. Um $\theta_1$ maior tende a diminuir o aprimoramento e, consequentemente, a carga viral.
*   **$\theta_{2}$ (relacionado a $N_0(A_2)$ - não neutralização):
** Um $\theta_{2}$ maior dificulta a neutralização do vírus pelos anticorpos, resultando em uma maior carga viral. Um $\theta_{2}$ menor favorece a neutralização e leva a uma menor carga viral.

```python
OgX = Omega/gamma_x

x, s = RungeKutta4(ModeloGeral, np.array([10.0, 0.0, OgX, 0.0, 100]), 0.001, 14)

# Converte o tempo de dias para meses (variável t_dias já é em dias)
t_dias = x

# Aplicar "clamping" (limite inferior artificial) para evitar erros numéricos
# Isto previne avisos do tipo RuntimeWarning ao calcular log10 ou potências
A1_sol_clamped = np.maximum(1e-10, s[:, 0])
A2_sol_clamped = np.maximum(1e-10, s[:, 1])
X_sol_clamped = np.maximum(1e-10, s[:, 2])
Y_sol_clamped = np.maximum(1e-10, s[:, 3])
V_sol_clamped = np.maximum(1e-10, s[:, 4])  # V é a 5ª variável (índice 4)

# Calcula as funções E(A1), N0(A2) e N(A2) para exibição gráfica
E_A1_plot = ((theta_1**(n-p)) * (A1_sol_clamped**p)) / ((theta_1**n) + (A1_sol_clamped**n))

N0_A2_plot = (theta_2**n) / (theta_2**n + A2_sol_clamped**n)
NA2_plot  = (A2_sol_clamped**n) / (theta_2**n + A2_sol_clamped**n)

# Cria um grid de 2x2 subgráficos
fig, axes = plt.subplots(2, 2, figsize=(18, 14)) # Tamanho da figura ajustado
fig.suptitle('Dinâmica do Modelo de Dengue', fontsize=16)

annotations_data = [] # Lista para armazenar dados para a legenda em markdown

# Gráfico 1: Anticorpos A1 e A2
axes[0, 0].plot(t_dias, np.log10(A1_sol_clamped + 10), '-b', label='$A_1$')
axes[0, 0].plot(t_dias, np.log10(A2_sol_clamped + 10), '-r', label='$A_2$')
axes[0, 0].set_xlabel('tempo (dias)')
axes[0, 0].set_ylabel(r'$	ext{log}_{10}(	ext{Anticorpos} + 10)$')
axes[0, 0].legend(loc=1)
axes[0, 0].set_title('Anticorpos $A_1$ e $A_2$')

# Max A1
max_idx_A1 = np.argmax(A1_sol_clamped)
t_max_A1 = t_dias[max_idx_A1]
val_max_A1_orig = A1_sol_clamped[max_idx_A1]
val_max_A1_plot = np.log10(val_max_A1_orig + 10)
axes[0, 0].plot(t_max_A1, val_max_A1_plot, 'o', color='blue', markersize=8, markeredgecolor='black')
annotations_data.append({'plot': 'Anticorpos $A_1$ e $A_2$', 'type': 'Máximo $A_1$', 'tempo': t_max_A1, 'valor_original': val_max_A1_orig, 'valor_plotado': val_max_A1_plot})

# Max A2
max_idx_A2 = np.argmax(A2_sol_clamped)
t_max_A2 = t_dias[max_idx_A2]
val_max_A2_orig = A2_sol_clamped[max_idx_A2]
val_max_A2_plot = np.log10(val_max_A2_orig + 10)
axes[0, 0].plot(t_max_A2, val_max_A2_plot, 'o', color='red', markersize=8, markeredgecolor='black')
annotations_data.append({'plot': 'Anticorpos $A_1$ e $A_2$', 'type': 'Máximo $A_2$', 'tempo': t_max_A2, 'valor_original': val_max_A2_orig, 'valor_plotado': val_max_A2_plot})

# Intersection A1 and A2
log_A1_plot = np.log10(A1_sol_clamped + 10)
log_A2_plot = np.log10(A2_sol_clamped + 10)
diff_log_A = log_A1_plot - log_A2_plot
intersections_A_indices = np.where(np.diff(np.sign(diff_log_A)))[0]

if len(intersections_A_indices) > 0:
    idx = intersections_A_indices[0]
    t_intersec_A = t_dias[idx]
    val_intersec_A_plot = log_A1_plot[idx] # Usar o valor de A1 no ponto de intersecção para plotagem
    val_intersec_A_orig = A1_sol_clamped[idx] # Usar o valor original de A1 no ponto de intersecção
    axes[0, 0].plot(t_intersec_A, val_intersec_A_plot, 'x', color='green', markersize=10, markeredgewidth=2)
    annotations_data.append({'plot': 'Anticorpos $A_1$ e $A_2$', 'type': 'Intersecção $A_1$ e $A_2$', 'tempo': t_intersec_A, 'valor_original': val_intersec_A_orig, 'valor_plotado': val_intersec_A_plot})


# Gráfico 2: Populações de Células X e Y
axes[0, 1].plot(t_dias, np.log10(X_sol_clamped/OgX + 10), '-b', label='$X$')
axes[0, 1].plot(t_dias, np.log10(Y_sol_clamped/OgX + 10), '-r', label='$Y$')
axes[0, 1].set_xlabel('tempo (dias)')
axes[0, 1].set_ylabel(r'$	ext{log}_{10}(	ext{Pop. Células} + 10)$')
axes[0, 1].legend(loc=1)
axes[0, 1].set_title('População de Células $X$ e $Y$')

# Max X (initial value is usually max for X)
max_idx_X = np.argmax(X_sol_clamped)
t_max_X = t_dias[max_idx_X]
val_max_X_orig = X_sol_clamped[max_idx_X]
val_max_X_plot = np.log10(val_max_X_orig/OgX + 10)
axes[0, 1].plot(t_max_X, val_max_X_plot, 'o', color='blue', markersize=8, markeredgecolor='black')
annotations_data.append({'plot': 'Populações de Células $X$ e $Y$', 'type': 'Máximo $X$', 'tempo': t_max_X, 'valor_original': val_max_X_orig, 'valor_plotado': val_max_X_plot})

# Max Y
max_idx_Y = np.argmax(Y_sol_clamped)
t_max_Y = t_dias[max_idx_Y]
val_max_Y_orig = Y_sol_clamped[max_idx_Y]
val_max_Y_plot = np.log10(val_max_Y_orig/OgX + 10)
axes[0, 1].plot(t_max_Y, val_max_Y_plot, 'o', color='red', markersize=8, markeredgecolor='black')
annotations_data.append({'plot': 'Populações de Células $X$ e $Y$', 'type': 'Máximo $Y$', 'tempo': t_max_Y, 'valor_original': val_max_Y_orig, 'valor_plotado': val_max_Y_plot})

# Gráfico 3: População Viral V
axes[1, 0].plot(t_dias, np.log10(V_sol_clamped + 10), '-r', label='$V$')
axes[1, 0].plot(t_dias, np.full_like(t_dias, np.log10(357 + 10)), '--k', label='Limite de detecção')
axes[1, 0].set_xlabel('tempo (dias)')
axes[1, 0].set_ylabel(r'$	ext{log}_{10}(V + 10)$')
axes[1, 0].legend(loc=1)
axes[1, 0].set_title('População Viral $V$')

# Max V
max_idx_V = np.argmax(V_sol_clamped)
t_max_V = t_dias[max_idx_V]
val_max_V_orig = V_sol_clamped[max_idx_V]
val_max_V_plot = np.log10(val_max_V_orig + 10)
axes[1, 0].plot(t_max_V, val_max_V_plot, 'o', color='red', markersize=8, markeredgecolor='black')
annotations_data.append({'plot': 'População Viral $V$', 'type': 'Máximo $V$', 'tempo': t_max_V, 'valor_original': val_max_V_orig, 'valor_plotado': val_max_V_plot})


# Gráfico 4: Funções E(A1) e N(A2)
axes[1, 1].plot(t_dias, E_A1_plot, '-r', label='E(A1)')
axes[1, 1].plot(t_dias, NA2_plot, '-b', label='N(A2)')
axes[1, 1].set_xlabel('tempo (dias)')
axes[1, 1].set_ylabel('Funções E e N')
axes[1, 1].legend(loc=1)
axes[1, 1].set_title('Funções E(A1) e N(A2)')

# Max E_A1
max_idx_E_A1 = np.argmax(E_A1_plot)
t_max_E_A1 = t_dias[max_idx_E_A1]
val_max_E_A1 = E_A1_plot[max_idx_E_A1]
axes[1, 1].plot(t_max_E_A1, val_max_E_A1, 'o', color='red', markersize=8, markeredgecolor='black')
annotations_data.append({'plot': 'Funções E(A1) e N(A2)', 'type': 'Máximo E(A1)', 'tempo': t_max_E_A1, 'valor_original': val_max_E_A1})

# Max NA2
max_idx_NA2 = np.argmax(NA2_plot)
t_max_NA2 = t_dias[max_idx_NA2]
val_max_NA2 = NA2_plot[max_idx_NA2]
axes[1, 1].plot(t_max_NA2, val_max_NA2, 'o', color='blue', markersize=8, markeredgecolor='black')
annotations_data.append({'plot': 'Funções E(A1) e N(A2)', 'type': 'Máximo N(A2)', 'tempo': t_max_NA2, 'valor_original': val_max_NA2})

# Intersection E(A1) and N(A2)
diff_E_NA2 = E_A1_plot - NA2_plot
intersections_E_NA2_indices = np.where(np.diff(np.sign(diff_E_NA2)))[0]

if len(intersections_E_NA2_indices) > 0:
    idx = intersections_E_NA2_indices[0]
    t_intersec_E_NA2 = t_dias[idx]
    val_intersec_E_NA2 = (E_A1_plot[idx] + NA2_plot[idx]) / 2 # Average for plotting
    axes[1, 1].plot(t_intersec_E_NA2, val_intersec_E_NA2, 'x', color='purple', markersize=10, markeredgewidth=2)
    annotations_data.append({'plot': 'Funções E(A1) e N(A2)', 'type': 'Intersecção E(A1) e N(A2)', 'tempo': t_intersec_E_NA2, 'valor_original': val_intersec_E_NA2})

# Ajusta layout para evitar sobreposição e deixar espaço para o título principal
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Prepare markdown table for annotations_data
markdown_table = "### Legenda dos Pontos Anotados\n\n"
markdown_table += "| Gráfico | Tipo de Ponto | Tempo (dias) | Valor Original | Valor Plotado (log10 + 10 / direto) |\n"
markdown_table += "|---|---|---|---|---|\n"

for item in annotations_data:
    plot_name = item['plot']
    point_type = item['type']
    time_val = f"{item['tempo']:.1f}"
    original_val = f"{item['valor_original']:.2e}"
    # For plots 1,2,3 y-axis is log-scaled. For plot 4, it's direct.
    if 'valor_plotado' in item:
        plot_val = f"{item['valor_plotado']:.2f}"
    else:
        plot_val = f"{item['valor_original']:.2f}" # Use original for non-log scaled plots
    markdown_table += f"| {plot_name} | {point_type} | {time_val} | {original_val} | {plot_val} |\n"

# Display the markdown table in a new cell
_ = markdown_table # Store in a dummy variable to avoid printing twice
```
